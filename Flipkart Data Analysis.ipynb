## 1. PROJECT TITLE : Flipkart Refrigerators Data Analysis

### Overview

This project involved extracting product data from the Flipkart website using web scraping techniques and performing exploratory data analysis (EDA) to understand pricing patterns, rating distribution, and key product features. Using Python libraries such as BeautifulSoup, Requests, and Pandas, I scraped product names, prices, ratings, reviews, and specifications across multiple pages. The collected data was cleaned, transformed, and analyzed to uncover insights around customer preferences and market trends. Visualizations were created to identify relationships between price, ratings, and product popularity.

## Project Objective

**The main objective of this project was to collect real-time e-commerce data from Flipkart and analyze it to derive actionable insights. Specific goals included:**

- Automating the extraction of product information from multiple pages.

- Cleaning and structuring the scraped data for analysis.

- Identifying pricing trends and rating patterns across product categories.

- Understanding factors influencing product popularity and customer choice.

- Building visualizations to communicate insights effectively.

## 1. Problem Statement

- E-commerce platforms like Flipkart contain large volumes of real-time product information, but this data is not readily available in a structured format for analysis. The challenge was to extract product details such as price, ratings, reviews, and specifications directly from Flipkart, clean and organize the data, and perform exploratory analysis to understand customer behavior and market trends.

## 2. Methodology 

1. Data Collection:

Web scraping is performed using requests and BeautifulSoup to extract from the flipkart website.

The scraped data is stored in a structured format using Pandas DataFrames.

2.Data Cleaning:

Remove redundant rows .

Convert appropriate columns to numerical data types.

Handle missing or inconsistent values.

3. Data Analysis and Visualization (EDA):

Analyze distributions of points, assists, rebounds, etc.

Use box plots, histograms, and heatmaps to understand trends and correlations.

Group data by positions and teams for comparative analysis.

4.Visualization:

Use matplotlib and seaborn to visualize key insights like top scorers, assist leaders, or team-wise average stats.

Correlation matrix to observe interdependencies between various stats.
